Explanation:
Data Preparation:
company, revenue, and profit lists contain the necessary data.

Setting Up the X-axis Positions:
ind = np.arange(n) creates an array of positions for the bars.
width = 0.35 sets the width of the bars.

Plotting Bars:
bar1 = ax.bar(ind, revenue, width, label='Revenue') plots the revenue bars.
bar2 = ax.bar(ind + width, profit, width, label='Profit') plots the profit bars next to the revenue bars.

Labels, Title, and Legend:
ax.set_xlabel('Company'), ax.set_ylabel('Amount (in billions)'), and ax.set_title('Revenue and Profit by Company') add the respective labels and title.
ax.set_xticks(ind + width / 2) and ax.set_xticklabels(company) set the x-axis ticks and labels.
ax.legend() adds a legend to differentiate between revenue and profit.
Display the Chart:

plt.show() displays the bar chart.
This code will produce a bar chart with revenue and profit bars side by side for each company.








```python
import matplotlib.pyplot as plt
import numpy as np

# Data
company = ['GOOGL', 'AMZN', 'MSFT', 'FB']
revenue = [90, 136, 89, 27]
profit = [40, 2, 34, 12]

# Number of companies
n = len(company)

# Create an array for the X-axis positions
ind = np.arange(n)

# Width of the bars
width = 0.35

# Plotting
fig, ax = plt.subplots()

# Bars for revenue
bar1 = ax.bar(ind, revenue, width, label='Revenue')

# Bars for profit
bar2 = ax.bar(ind + width, profit, width, label='Profit')

# Adding labels, title, and legend
ax.set_xlabel('Company')
ax.set_ylabel('Amount (in billions)')
ax.set_title('Revenue and Profit by Company')
ax.set_xticks(ind + width / 2)
ax.set_xticklabels(company)
ax.legend(shadow=True)

# Display the chart
plt.show()
```
