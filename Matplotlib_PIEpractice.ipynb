CREATING A PIE CHART

```python
import matplotlib.pyplot as plt
```

plot a pie chart to track down home expenses

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.pie(exp_vals,labels=exp_labels, shadow=True)
plt.show()
```

Show percentages for every pie. Specify radius to increase chart size

```python
plt.axis("equal") #use 'equal' to make the pie chart even
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%0.1f%%',radius=1.5) #use autopct to show percentages of every pie
plt.show()                                                                      #0.1 for one decimal points and 0.2 for 2dp
```

Explode

```python
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.1f%%',radius=1.5,explode=[0,0,0,0.1,0.2]) #can change each value according to pie that you wanna explode
plt.show()
```

counterclock and change pie angles

```python
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.1f%%',radius=1.5,explode=[0,0,0,0.1,0.2],counterclock=True, startangle=45)
plt.show()
```
