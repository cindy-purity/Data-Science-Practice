Import pandas library

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/Most Streamed Spotify Songs 2024.csv', encoding = 'latin-1')
```

Inspect the data

```python
df.head()
```

View the columns the DataFrame

```python
df.columns
```

Check for null Values in the DataFrame

```python
df.isnull().sum()
```

Replace all null values with Zero

```python
df.fillna(0,inplace=True)
```

```python
df.isnull().sum()
```

Check and drop all duplicates

```python
df.duplicated().sum()
```

```python
df[df.duplicated()]
```

```python
df.drop_duplicates(inplace=True)
```

```python
#drop multiple columns
df.drop(['ISRC', 'Track Score','YouTube Views','TikTok Views',
       'Spotify Playlist Count', 'Spotify Playlist Reach',
       'YouTube Likes', 'TikTok Posts',
       'TikTok Likes', 'YouTube Playlist Reach',
       'Apple Music Playlist Count', 'AirPlay Spins', 'SiriusXM Spins',
       'Deezer Playlist Count', 'Deezer Playlist Reach',
       'Amazon Playlist Count', 'Pandora Streams', 'Pandora Track Stations',
       'Soundcloud Streams', 'Shazam Counts', 'TIDAL Popularity','Spotify Popularity','Explicit Track'], axis = 1, inplace = True)
```

```python
df.columns
```

```python
df.head(5)
```

Access the Spotify streams columns

```python
df_spotify_streams = df[['Track','Spotify Streams']].sort_values(by='Spotify Streams', ascending = False)
df_spotify_streams.head(20)
```

```python
# Convert the 'Track' column to strings, handling potential errors
df_spotify_streams['Track'] = df_spotify_streams['Track'].apply(lambda x: str(x) if not pd.isna(x) else '')
```

Import matplotlib library

```python
import matplotlib.pyplot as plt
```

Visualize the Data

```python
#plot a figure of 10 most streamed songs on spotify
plt.figure(figsize=(10, 6))
plt.title('Top 10 Most Streamed Songs on Spotify')
plt.xlabel('Song')
plt.ylabel('Spotify Streams')
plt.bar(df_spotify_streams['Track'].head(10), df_spotify_streams['Spotify Streams'].head(10))
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
top_20_artist = df.sort_values(by='Spotify Streams',ascending=False).head(100).value_counts('Artist').head(20)
top_20_artist
```

```python
# 20 most streamed artist in 2024
plt.figure(figsize=(10, 6))
plt.title('Top 20 Most Streamed Artists on Spotify')
plt.xlabel('Artist')
plt.ylabel('no of tracks')
plt.bar(top_20_artist.index, top_20_artist.values)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
